### Trabajos con valores nulos

```python
#Validar si existe valores nulos
df.isnull().sum()
```

```python
#Identificar registros unicos del campo nivel_estudios
print(df['nivel_estudios'].unique())
```

```python
#Evaluar si reemplazamos valores nulos o los eliminados
df[df['nivel_estudios'].isnull()]# seleccionamos las filas con valores nulos en nivel_estudios
```

```python
#Para no perder informacion reemplazaremos los valores nulos por sin información

df['nivel_estudios'] = df['nivel_estudios'].fillna('Sin información')

#Validamos  el cambio
print(df['nivel_estudios'].unique())
```

#### Trabajos nulos con columna satisfacción laboral

```python
#Identificar registros unicos del campo satisfaccion_laboral
print(df['satisfaccion_laboral'].unique())
```

```python
#Evaluar si reemplazamos valores nulos o los eliminados
df[df['satisfaccion_laboral'].isnull()]# seleccionamos las filas con valores nulos en satisfacción laboral 
```

```python
#Para no perder informacion reemplazaremos los valores nulos por sin información
df['satisfaccion_laboral'] = df['satisfaccion_laboral'].fillna(df['satisfaccion_laboral'].mean())

#Validamos  el cambio
print(df['satisfaccion_laboral'].unique())
```

### Analisis con el Pais Chile

```python
#Filtrar la data Pais = Chile y la almacenamos en una nueva dataframe

df_chile = df_filtrado = df[df['pais'] == 'Chile']

df_chile.head(15)
```

```python
#Calcularemos la media, mediana y desviación estándar de los sueldos.

df_chile['sueldo_mensual_usd'].describe()
```

```python
media_sueldo_chile = df_chile['sueldo_mensual_usd'].mean()
mediana_sueldo_chile = df_chile['sueldo_mensual_usd'].median()
desviacion_sueldo_chile = df_chile['sueldo_mensual_usd'].std()

print(f"Media: {media_sueldo_chile:.2f}")
print(f"Mediana: {mediana_sueldo_chile:.2f}")
print(f"Desviación estándar: {desviacion_sueldo_chile:.2f}")
```
